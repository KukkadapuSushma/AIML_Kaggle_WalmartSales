```python
from pandas import read_csv
from datetime import datetime
def labelConvert(s):
    s = s.strip().lower()
    if s == "womenclothing":
        return 0
    if s == "menclothing":
        return 1
    return 2

dataset = read_csv('Train_Kaggle.csv', converters={2:labelConvert})
#del dataset['Sales(In ThousandDollars)']
dataset
```

```python
from pandas import read_csv
dataset_test = read_csv('Test_Kaggle.csv', converters={2:labelConvert})
del dataset_test['Sales(In ThousandDollars)']
dataset_test
```

```python

# manually specify column names
dataset['Sales(In ThousandDollars)'].fillna(0, inplace=True)
#print(dataset)
dataset.to_csv('sales.csv')

data_Y_trainsales = dataset[['Sales(In ThousandDollars)']]
data_X_train = dataset.iloc[:,0:7]
del data_X_train['Sales(In ThousandDollars)']
#data_Y_trainsales
data_X_train
```
